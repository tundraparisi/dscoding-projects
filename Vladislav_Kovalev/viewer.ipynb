```python
dataHotels = pd.read_excel(dataHotelsPathWithoutGlobalRepo)
dataGuests = pd.read_excel(dataGuestsPathWithoutGlobalRepo)
dataPreferences = pd.read_excel(dataPreferencesPathWithoutGlobalRepo)
```

```python
dataFirstStrategy = pd.read_excel(firstStrategyPathWithoutGlobalRepo)
dataSecondStrategy = pd.read_excel(secondStrategyPathWithoutGlobalRepo)
dataThirdStrategy = pd.read_excel(thirdStrategyPathWithoutGlobalRepo)
dataFourthStrategy = pd.read_excel(fourthStrategyPathWithoutGlobalRepo)
```

```python
strategies = ['First Strategy', 'Second Strategy', 'Third Strategy', 'Fourth Strategy']
netto_sum = [
    dataFirstStrategy.NETTO.sum(),
    dataSecondStrategy.NETTO.sum(),
    dataThirdStrategy.NETTO.sum(),
    dataFourthStrategy.NETTO.sum()
]
colors = ['steelblue', 'orange', 'green', 'purple']

plt.bar(strategies, netto_sum, color=colors)

plt.xlabel('Strategies')
plt.ylabel('Netto Sum')
plt.title('Netto Sum by Strategy')

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
bins = 100

# Strategy 1
axes[0, 0].set_xlabel('price')
axes[0, 0].set_ylabel('number of guests')
axes[0, 0].hist(dataFirstStrategy.NETTO, bins=bins, color='steelblue')
axes[0, 0].set_title('Histogram - Strategy 1')

mean_1 = np.mean(dataFirstStrategy.NETTO)
mode_1 = dataFirstStrategy.NETTO.mode().mean()
median_1 = dataFirstStrategy.NETTO.median()

axes[0, 0].axvline(x=mean_1, color='r', linestyle='--', label='Mean')
axes[0, 0].axvline(x=mode_1, color='g', linestyle='--', label='Mode')
axes[0, 0].axvline(x=median_1, color='b', linestyle='--', label='Median')
axes[0, 0].legend()

# Strategy 2
axes[0, 1].set_xlabel('price')
axes[0, 1].set_ylabel('number of guests')
axes[0, 1].hist(dataSecondStrategy.NETTO, bins=bins, color='orange')
axes[0, 1].set_title('Histogram - Strategy 2')

mean_2 = np.mean(dataSecondStrategy.NETTO)
mode_2 = dataSecondStrategy.NETTO.mode().mean()
median_2 = dataSecondStrategy.NETTO.median()

axes[0, 1].axvline(x=mean_2, color='r', linestyle='--', label='Mean')
axes[0, 1].axvline(x=mode_2, color='g', linestyle='--', label='Mode')
axes[0, 1].axvline(x=median_2, color='b', linestyle='--', label='Median')
axes[0, 1].legend()

# Strategy 3
axes[1, 0].set_xlabel('price')
axes[1, 0].set_ylabel('number of guests')
axes[1, 0].hist(dataThirdStrategy.NETTO, bins=bins, color='green')
axes[1, 0].set_title('Histogram - Strategy 3')

mean_3 = np.mean(dataThirdStrategy.NETTO)
mode_3 = dataThirdStrategy.NETTO.mode().mean()
median_3 = dataThirdStrategy.NETTO.median()

axes[1, 0].axvline(x=mean_3, color='r', linestyle='--', label='Mean')
axes[1, 0].axvline(x=mode_3, color='g', linestyle='--', label='Mode')
axes[1, 0].axvline(x=median_3, color='b', linestyle='--', label='Median')
axes[1, 0].legend()

# Strategy 4
axes[1, 1].set_xlabel('price')
axes[1, 1].set_ylabel('number of guests')
axes[1, 1].hist(dataFourthStrategy.NETTO, bins=bins, color='purple')
axes[1, 1].set_title('Histogram - Strategy 4')

mean_4 = np.mean(dataFourthStrategy.NETTO)
mode_4 = dataFourthStrategy.NETTO.mode().mean()
median_4 = dataFourthStrategy.NETTO.median()

axes[1, 1].axvline(x=mean_4, color='r', linestyle='--', label='Mean')
axes[1, 1].axvline(x=mode_4, color='g', linestyle='--', label='Mode')
axes[1, 1].axvline(x=median_4, color='b', linestyle='--', label='Median')
axes[1, 1].legend()


axes[0, 0].text(0.05, 0.9, f'Mode: {mode_1:.2f}\nMedian: {median_1:.2f}',
                transform=axes[0, 0].transAxes, fontsize=10)
axes[0, 1].text(0.05, 0.9, f'Mode: {mode_2:.2f}\nMedian: {median_2:.2f}',
                transform=axes[0, 1].transAxes, fontsize=10)
axes[1, 0].text(0.05, 0.9, f'Mode: {mode_3:.2f}\nMedian: {median_3:.2f}',
                transform=axes[1, 0].transAxes, fontsize=10)
axes[1, 1].text(0.05, 0.9, f'Mode: {mode_4:.2f}\nMedian: {median_4:.2f}',
                transform=axes[1, 1].transAxes, fontsize=10)

fig.tight_layout()
plt.show()
```
