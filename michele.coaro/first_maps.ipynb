```python
import folium as f
import pandas as pd
```

```python
ctWdates = pd.read_csv("DFtoImport.csv")
uniqueLoc = pd.read_csv("Unique_cities_coordinates.csv")
```

## Creating a folium Map

```python
ctWdates.head()
```

```python
uniqueLoc.head()
```

```python
max_lat = max(ctWdates['Latitude']+2)
min_lat = min(ctWdates['Latitude']-2)
max_lon = max(ctWdates['Longitude']+2)
min_lon = min(ctWdates['Longitude']+2)
```

```python
background = f.Figure (width = 2000, height = 1500)
mapp = f.Map(location = (30,10), zoom_start = 2.2, tiles = ("cartodb positron"), max_bounds = True, control_scale = True, minZoom = 2)


# instantiate a feature group for the incidents in the dataframe
Locations = f.map.FeatureGroup()
pop = 0
# loop through the 100 crimes and add each to the incidents feature group
for lat, lon, in zip(uniqueLoc.Lat, uniqueLoc.Lon):
    Locations.add_child(
        f.features.CircleMarker(                
            [lat, lon],
            radius=5, # define how big you want the circle markers to be
            color='blue',
            fill=True,
            fill_color='yellow',                
            fill_opacity=0.5,
            popup=(uniqueLoc['City'][pop]),
            
        )
    )
    pop +=1
    
mapp.add_child(Locations)

mapp.fit_bounds((max_lat, max_lon), (min_lat, min_lon))
mapp.save('mappa_solo_città.html')
mapp
```
